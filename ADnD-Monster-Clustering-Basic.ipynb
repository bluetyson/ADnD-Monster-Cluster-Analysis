```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
km = pd.read_csv(r'D:\Dungeon23\RandomDungeon\kmodes clusters.csv')
fig, ax = plt.subplots()
km['Frequency'] = km['Frequency'].astype('str')
km.plot.scatter('Frequency','Type',c='adnd$cluster',cmap='Spectral',ax=ax,figsize=(21,10),s=600,marker='h')
ax.tick_params(top=False,
               bottom=False,
               left=False,
               right=False,
               labelleft=True,
               labelbottom=True)
plt.xticks(rotation=45, ha='right')
plt.show()
```
